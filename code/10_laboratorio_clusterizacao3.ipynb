```python
dados_noticias = pd.read_csv("../datasets/clusterizacao/DataNews.csv")
dados_noticias.dropna(inplace=True)
dados_noticias.head()
```

```python
dados_noticias.shape
```

```python
# Imports necessários
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
arquivo = open('../datasets/clusterizacao/stopwordsPtBr.txt', 'r')
arquivo

stopwords = []

for linha in arquivo:
  valores = linha.strip()
  stopwords.append(valores)

# stopwords
```

```python
# Criação da lista de textos que serão vetorizados. O método não aceita o dataframe como entrada. 
texts = list(dados_noticias['text'])

# Aplicação da Vetorização
tfidf_vectorizer = TfidfVectorizer(max_df=0.8,
                                 min_df=0.2,
                                 use_idf=True, ngram_range=(1,3),
                                 stop_words=stopwords
                                            
                                 )
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)
```

A atividade consiste em aplicar técnicas de clusterização para agrupar os documentos e analise a partir dos textos agrupados de qual categoria consiste cada um dos grupos de notícias. 

Divirtam-se ;)

```python
# Variando o número de k de 1 a 10 para analisar os valores da métrica

for k in range(2, 10):
    km = KMeans(n_clusters=k,random_state=42)
    km.fit(tfidf_matrix)
    score = metrics.silhouette_score(tfidf_matrix, km.labels_, metric='euclidean')
    print("Para k=%i o valor Silhoutte foi de %.4f" % (k, score))
```

```python
km = KMeans(n_clusters=7,random_state=42)
clusters_km = km.fit_predict(tfidf_matrix)
```

```python
dados_noticias['cluster_km'] = clusters_km
dados_noticias.head(40)
```

```python
dados_noticias['cluster_km'].value_counts()
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
```

```python
# Criando e plotando o dendrograma
# plt.figure(figsize=(10, 7))
# plt.title("Dendograma Noticias")
# cluster_alg = shc.linkage(tfidf_matrix.toarray(), method='ward',)
# dend = shc.dendrogram(cluster_alg)
```

```python
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
order_centroids
```

```python
# Criação de uma veriável com a lista de palavras que representa as colunas da matriz.
terms = tfidf_vectorizer.get_feature_names_out()
```

```python
centroids = []
# 5 foi o número de clusters que usamos para treinar o modelo do K-Means
for i in range(7): 
  list_id = order_centroids[i]
  list_terms = [terms[id] for id in list_id[:10]]
  centroids.append({'cluster': i, 'words': ', '.join(list_terms)})

df_centroids = pd.DataFrame(centroids)
df_centroids
```

```python
for n in range(7):
  print("Cluster %i: %s" % (n, list(df_centroids[df_centroids['cluster'] == n]['words'][:5])))
  data_selected = dados_noticias[dados_noticias['cluster_km'] == n]
  for index, row in data_selected.head(10).iterrows():
    print("\tNoticia: %s" % row['text'])
  print()
```
